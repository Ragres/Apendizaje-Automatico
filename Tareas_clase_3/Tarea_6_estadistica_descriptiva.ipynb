```python
#Se asume que el gasto total sigue una distribucion gaussania y se hace la prueba de normalidad para comprobar

from scipy.stats import shapiro
stat, p = shapiro(Datos["Total expenditure"])
alpha = 0.05
if p < alpha:
    print(f'La distribución no parece normal con p = {round(p, 4)}')
else:
    print(f'La distribución parece normal con p = {round(p, 4)} ')
```

```python
#Se revisa el qq plot para comprobar lo estimado
from statsmodels.graphics.gofplots import qqplot

plt.figure()
qqplot(Datos["Total expenditure"], line='s')
plt.show()
```

```python
def obtener_media(lista):
    media=sum(lista)/len(lista)
    return(media)

def obtener_varianza(lista):
    lista=[x-obtener_media(lista) for x in lista]
    lista=[x**2 for x in lista]
    varianza=sum(lista)/(len(lista)-1)
    return(varianza)
```

```python
#Comprobando si mi variable de interes (Life expectancy) es parametrica o no
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.figure()
plt.hist(Datos["Total expenditure"], density=True)
mu=obtener_media(Datos["Total expenditure"])
sigma=obtener_varianza(Datos["Total expenditure"])
plt.plot(np.arange(0,20, .1), norm.pdf(np.arange(0,20,.1), mu, sigma), label = 'Curva gaussiana')
plt.legend()
plt.show()
```

```python
#Si bien la distribucion puede parecer normal se aprecia un sesgo concentrado al centro el cual desvia los cuantiles
```
